```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

print(menMeans)
print(womenMeans)

# visualizaÃ§Ã£o "empilhada"
p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

```python
import numpy as np
import seaborn as sns
sns.set(style="ticks")

rs = np.random.RandomState(11)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)

sns.jointplot(x, y, kind="hex", color="#4CB391")
```

```python
corr = np.array([1,0,1,0,1,0,1,0,1]).reshape(3,3)
sns.heatmap(data=corr, annot=True)
```

```python
classe_a = np.random.normal(50, 10, size=20)
classe_b = np.random.normal(70, 10, size=20)

plt.hist(classe_a, bins=range(0,100,10), alpha=1.0)
plt.hist(classe_b, bins=range(0,100,10), alpha=0.5)
plt.show()
```

```python
ax = sns.distplot(classe_a, bins=range(0,100,10), kde=False, norm_hist=False)
ax = sns.distplot(classe_b, bins=range(0,100,10), kde=False, norm_hist=False)
```

```python
ax = sns.distplot(classe_a, bins=range(0,100,10))
ax = sns.distplot(classe_b, bins=range(0,100,10))
```

```python
import pandas as pd
classes = len(classe_a)*['A'] + len(classe_b)*['B']
notas = np.concatenate((classe_a, classe_b))

df = pd.DataFrame({'Notas': notas, 'Turma': classes})
# print(df)
sns.boxplot(x="Turma", y="Notas", data=df)
```

### Visualizando atributos do conjunto Iris

```python
import matplotlib.pyplot as plt

from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
colors=np.array(50*['r']+50*['g']+50*['b'])
pd.plotting.scatter_matrix(df, alpha=1.0, figsize=(15,15), color=colors)
```

```python
df = sns.load_dataset('iris')
sns.pairplot(df, hue="species")
```

```python
plt.scatter(df['petal_length'], df['petal_width'])
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].scatter(df['sepal_length'], df['petal_length'], c=colors)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].scatter(df['sepal_length'], df['sepal_width'], c=colors)
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].scatter(df['sepal_length'], df['petal_length'], c=colors)
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].scatter(df['sepal_length'], df['petal_width'], c=colors)
axs[1, 1].set_title('Axis [1, 1]')
```

### Pie chart

```python
letters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
dist = [20, 20, 5, 5, 3, 7, 10, 20, 10]

plt.pie(dist, labels=letters)
```

```python
new_dist = [0]
letras = ['outros']

for l,v in zip(letters, dist):
  if v < 20:
    new_dist[0] += v
  else:
    letras.append(l)
    new_dist.append(v)

plt.pie(new_dist, labels=letras)
```

```python
plt.pie(new_dist, labels=letras, autopct='%.2f')
```
